```python
# 1. Car Pooling
def carPooling(trips, capacity):
    if not trips:
        return False

    city_num = max(max(x[1:]) for x in trips)

    # a list of the city traveling through
    loc = list(range(city_num))
    # a list of vacant spot in each city
    pas = [0 for _ in range(city_num)]

    # fill passengers into the pas list
    trip = []
    for t in trips:
        trip_i = pas[0:t[1]-1] + [t[0] for _ in range(t[1]-1, t[2]-1)] + pas[t[2]-1:]
        trip.append(trip_i)

    pas = [sum(x) for x in zip(*trip)]
    print(pas)

    if max(pas) > capacity:
        return False
    else:
        return True


carPooling([[2,1,5],[3,5,7]], 4)
```

```python
# 3. Pancake Sorting
def pancakeSort(arr):
  """
  :type arr: List[int]
  :rtype: List[int]
  """
  def findMaxIdx(arr):
    return arr.index(max(arr))

  def reverseSubArr(arr, k):
    return list(reversed(arr[:k+1])) + arr[k+1:]
  
  arr_len = len(arr)
  arr_ = arr.copy()
  output = []
  while arr_len > 1:
    arr_ = arr_[:arr_len]
    print("arr_len = {}, arr_ = {}".format(arr_len, arr_))
    
    k = findMaxIdx(arr_)
    if k != arr_len:
      output.append(k+1)
      # largest element is flipped to arr_[0]
      arr_ = reverseSubArr(arr_, k)
      print("k = {}, arr_ = {}".format(k, arr_))

      k = len(arr_) - 1
      output.append(k+1)
      arr_ = reverseSubArr(arr_, k)  # flip the whole arr_
      print("k = {}, arr_ = {}".format(k, arr_))

    arr_len -= 1
    
  return output

pancakeSort([3, 2, 4, 1])
# Output: [4,2,4,3]
# After 1st flip(k=4): arr = [1, 4, 2, 3]
# After 2nd flip(k=2): arr = [4, 1, 2, 3]
# After 3rd flip(k=4): arr = [3, 2, 1, 4]
# After 4th flip(k=3): arr = [1, 2, 3, 4]
```

```python
# 4. Optimal Division
def optimalDivision(nums):
    """
    :type nums: List[int]
    :rtype: str
    """
    nums_len = len(nums)
    if nums_len == 0:
      return ""
    
    if nums_len == 1:
      return str(nums[0])
    
    if nums_len == 2:
      return str(nums[0]) + "/" + str(nums[1])

    expression = ""
    for i in range(len(nums)):
      if i == 0:
        expression += str(nums[i])
        continue
        
      if i == 1:
        expression += "/(" + str(nums[i]) 
        continue
      
      if i < nums_len - 1:
        expression += "/" + str(nums[i])
        continue

      if i == nums_len - 1:
        expression += "/" +str(nums[i]) + ")"
        return expression




optimalDivision([1000, 100, 10, 2])
```

```python
# 5. Longest Continuous Increasing Subsequence
def findLengthOfLCIS(nums):
    """
    :type nums: List[int]
    :rtype: int
    """
    lcis = 1
    nums = nums + ["stop"]
    max_lcis = float("-inf")

    def isDigit(n):
      try:
        int_n = int(n)
        return True
      except:
        return False

    for i in range(len(nums)):
      num = nums[i]
      next_num = nums[i+1]
      if isDigit(next_num) and num < next_num:
        lcis += 1
      else:
        if lcis > max_lcis:
          max_lcis = lcis
        lcis = 1
      if not isDigit(next_num):
        break
    
    return max_lcis



findLengthOfLCIS([1,3,5,4,7])
# findLengthOfLCIS([2,2,2,2,2])
# findLengthOfLCIS(
    # [1, 3, 5, 4, 2, 3, 4, 5])
```

```python
# 6. Check If String Is a Prefix of Array
def isPrefixString(s, words):
    """
    :type s: str
    :type words: List[str]
    :rtype: bool
    """
    isps = False
    i = 0
    sw = ""
    while i < len(words):
      w = words[i]
      sw += w
      if sw == s:
        isps = True
      
      i += 1

    return isps



isPrefixString(s="iloveleetcode", words=["i", "love", "leetcode", "apples"])
# isPrefixString(s="iloveleetcode", words=["apples", "i", "love", "leetcode"])
```

```python
# 7. Lucky Numbers in a Matrix
def luckyNumbers(matrix):
  """
  :type matrix: List[List[int]]
  :rtype: List[int]
  """
  lucky_nums = []
  for j in range(len(matrix[0])):
    lucky = []
    col = []
    for row in matrix:
      col.append(row[j])
      lucky.append(min(row))
    
    if max(col) in lucky:
      lucky_nums.append(max(col))
    
  return lucky_nums



luckyNumbers([[3, 7, 8], 
              [9, 11, 13], 
              [15, 16, 17]])
# Output: [15]
luckyNumbers([[1, 10, 4, 2], 
              [9, 3, 8, 7], 
              [15, 16, 17, 12]])
# Output: [12]
luckyNumbers([[7, 8], 
              [1, 2]])
# Output: [7]
```

```python
# 8. Watering Plants
def wateringPlants(plants, capacity):
  """
  :type plants: List[int]
  :type capacity: int
  :rtype: int
  """
  step = 0
  water_level = capacity
  for i in range(len(plants)):
    step += 1
    water_level -= plants[i]

    if i != len(plants)-1 and water_level < plants[i+1]:
      # refilling water
      step += (i+1)
      water_level = capacity
      step += (i+1)

  return step


wateringPlants(plants=[2, 2, 3, 3], capacity=5)
# Output: 14
wateringPlants(plants=[1, 1, 1, 4, 2, 3], capacity=4)
# Output: 30
wateringPlants(plants=[7, 7, 7, 7, 7, 7, 7], capacity=8)
# Output: 49
```
